```python
from scipy.optimize import differential_evolution
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def func(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5.0 * np.exp(-x / 2.0)

bounds = [(1, 30)]
res = differential_evolution(func, bounds)
f1 = round(func(res.x),2)
print f1

X = np.linspace(1, 30, 256)
Y = func(X)

line0, = plt.plot(X, Y, color="black", linewidth=2.0, linestyle="-", label="f(x)")
line_min = plt.plot((res.x, res.x), (-12, 4), 'r-')
plt.axis((0,31,-12,4))
plt.show()

file_x_min = open("submission-2.txt", "w")
file_x_min.write(str(f1))
file_x_min.close()
```
