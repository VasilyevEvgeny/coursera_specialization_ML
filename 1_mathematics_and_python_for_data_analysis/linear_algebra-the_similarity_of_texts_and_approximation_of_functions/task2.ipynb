```python
import scipy.linalg
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def func(x):
    return np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)

# =============================================

x1 = 1
x2 = 15
A_p1 = np.array(([1, x1], [1, x2]))
b_p1 = np.array([func(x1), func(x2)])

x_p1 = scipy.linalg.solve(A_p1, b_p1)

def func_p1(x):
    return x_p1[0] + x_p1[1] * x


# =============================================

x1 = 1
x2 = 8
x3 = 15
A_p2 = np.array(([1, x1, x1**2], [1, x2, x2**2], [1, x3, x3**2]))
b_p2 = np.array([func(x1), func(x2), func(x3)])

x_p2 = scipy.linalg.solve(A_p2, b_p2)

def func_p2(x):
    return x_p2[0] + x_p2[1] * x + x_p2[2] * x**2

# =============================================

x1 = 1
x2 = 4
x3 = 10
x4 = 15
A_p3 = np.array(([1, x1, x1**2, x1**3], [1, x2, x2**2, x2**3], [1, x3, x3**2, x3**3], [1, x4, x4**2, x4**3]))
b_p3 = np.array([func(x1), func(x2), func(x3), func(x4)])

x_p3 = scipy.linalg.solve(A_p3, b_p3)
print x_p3

file_coeff = open("submission-2.txt", "w")
file_coeff.write(str(round(x_p3[0],2)) + ' ' + str(round(x_p3[1],2)) + ' ' \
                 + str(round(x_p3[2],2)) + ' ' + str(round(x_p3[3],2)))
file_coeff.close()
    
def func_p3(x):
    return x_p3[0] + x_p3[1] * x + x_p3[2] * x**2 + x_p3[3] * x**3

# =============================================

X = np.linspace(1, 15, 256)
Y = func(X)
Y1 = func_p1(X)
Y2 = func_p2(X)
Y3 = func_p3(X)

line0, = plt.plot(X, Y, color="black", linewidth=2.0, linestyle="-", label="f(x)")
line1, = plt.plot(X, Y1, color="blue", linewidth=1.0, linestyle=":", label="f(x)=O(x)")
line2, = plt.plot(X, Y2, color="red", linewidth=1.0, linestyle=":", label="f(x)=O(x**2)")
line3, = plt.plot(X, Y3, color="green", linewidth=1.0, linestyle=":", label="f(x)=O(x**3)")
legend = plt.legend(handles=[line0, line1, line2, line3])
plt.show()
```
